## Test Configurations

```python
from unittest.mock import MagicMock

# 1. The Stub: Providing data (Battery Level)
battery_stub = MagicMock()
battery_stub.get_level.return_value = 20  # Canned response

# 2. The Mock: Verifying side effects (Robot)
robot_mock = MagicMock()

class ControlSoftware:
    def __init__(self, battery, robot):
        self.battery = battery
        self.robot = robot
        
    def execute_command(self, command):
        # Using the Stub
        if self.battery.get_level() > 10:
            # Using the Mock
            self.robot.send_can_cmd(command)
            return "Success"
        return "Low Battery"

control_sw = ControlSoftware(battery_stub, robot_mock)
```

### Given the Battery Manager is reporting sufficient charge
<table style="text-align: left; margin-left: 0; margin-right: auto; width: auto; border-collapse: collapse;">
  <tbody><tr><td style="padding: 8px;">01</td><td style="padding: 8px;">Initialize system with Battery Manager stub.</td></tr></tbody>
</table>

```python
assert battery_stub.get_level() == 20
```

### When the UI sends 'DRIVE_FORWARD'
<table style="text-align: left; margin-left: 0; margin-right: auto; width: auto; border-collapse: collapse;">
  <tbody><tr><td style="padding: 8px;">02</td><td style="padding: 8px;">Call execute_command('DRIVE_FORWARD').</td></tr></tbody>
</table>

```python
result = control_sw.execute_command('DRIVE_FORWARD')
assert result == "Success"
```

### Then the Robot Mock should receive the command
<table style="text-align: left; margin-left: 0; margin-right: auto; width: auto; border-collapse: collapse;">
  <tbody><tr><td style="padding: 8px;">03</td><td style="padding: 8px;">Verify robot_mock received the correct CAN command.</td></tr></tbody>
</table>

```python
robot_mock.send_can_cmd.assert_called_with('DRIVE_FORWARD')
```
